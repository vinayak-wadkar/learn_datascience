# Data Visualization using python libraries ( Matplotlib and Seaborn)

#### 1) Matplotlib is a widely used and simple data visualization library with powerful features, many of the other libraries for data visualization are built upon matplotlib and is used as an extension to it, for more elaborate and detailed visualizations.

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

#### Basic Plot

```python
x = np.linspace(5,25,3)
```

```python
y = np.linspace(7,84,3) # dimension of both arrays should be same for linspace as it would throw an error if not.
```

```python
plt.plot(x,y)
```

```python
# need to call plt.show() explicitly to display the plot
plt.show()
```

```python
# can also work with lists, though it converts lists into numpy arrays internally
plt.plot([1, 4, 7, 10], [2, 3, 8, 13])
plt.show()
```

#### Let's see how to put labels and the x and y axes and the chart title.

#### Also, we can specify the limits of the x and y labels as a range using xlim([xmin, xmax]) and ylim([ymin, ymax])

```python
# Axis label and Title
plt.plot(x, y)

# x and y labels, and title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

# Define the range of the labels of the axis
# Arguments:
plt.xlim([0, 35])
plt.ylim([0, 100])
plt.show()
```

#### Customizing features of the visualization

```python
# initializing x and y arrays
x = np.linspace(0, 10, 20)
y = x * 2

# Colour for the plot -> blue which is denoted by 'b' and line type will be a '+' symbol
plt.plot(x, y, 'b+')

# x and y labels, and title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

plt.show()
```

#### Plotting multiple lines on the same graph

```python
x = np.linspace(0, 5 ,10)
y = np.linspace(3, 6, 10)

# Plttong three curves : y, y**2 and y**3 with different line types and colours
plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g*')
plt.show()
```

### Figures and Subplots
You often need to create multiple plots in the same figure, as we see in the example

matplotlib has the concept of figures and subplots using which you can create multiple subplots called Facets inside the same figure.

To create these facets in the same figure, you can use the method plt.subplots(nrows, ncols, nsubplot)

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

# initiate a new figure explicitly
plt.figure(1)

# Create a subplot with 1 row, 2 columns

# Create the first subplot in figure 1
plt.subplot(121)     # equivalent to plt.subplot(1, 2, 1)
plt.title("y = log(x)")
plt.plot(x, y)

# Create the second subplot in figure 1
plt.subplot(1, 2, 2)
plt.title("y = log(x)**2")
plt.plot(x, y**2)

plt.show()
```

#### Create a plot with four subplots (two rows and two columns)

```python
x = np.linspace(1, 10, 100)

# Optional command, since matplotlib creates a figure by default anyway, no need for explicitly calling this command
# plt.figure(1)

# Subplot 1
plt.subplot(2, 2, 1)
plt.title("Linear")
plt.plot(x, x)

# Subplot 2
plt.subplot(2, 2, 2)
plt.title("Exponential")
plt.plot(x**2)

#Subplot 3
plt.figure(2)
plt.subplot(2, 2, 1)
plt.title("Cubic")
plt.plot(x**3)

# Subplot 4
plt.subplot(2, 2, 2)
plt.title("Log")
plt.plot(np.log(x))

plt.show()
```

### Types of Commonly used plots
Let's now use the retail stores data to create some commonly used plots such as:

Boxplots,
Histograms,
Scatter plots,
Bar plots

```python
# Example Global salaes data
df = pd.read_csv("/home/vinayak/vDataScience/Datasets/Datavisual/global_sales_data/market_fact.csv")
df.head()
```

## Boxplots

```python
# Boxplot: Visualize the distribution of a continuous variable
plt.boxplot(df["Order_Quantity"])
plt.show()
```

```python
# Boxplot for Sales
plt.boxplot(df["Sales"])
plt.show()
```

As we can see, the boxplot of Sales is unreadable, since Sales varies greatly across a wide range

```python
df['Sales'].describe()
```

The solution to this problem is to change the scale of the axis (in this case, the y-axis) so that the range can fit into the size of the plot.

One commonly used technique is to transform an axis into the logarithmic scale. You can transform the scale of an axis using plt.yscale('log').

```python
# usual Linear scale subplot
plt.subplot(1, 2, 1)
plt.boxplot(df['Sales'])
plt.title('Normal Linear Scale')

# Log scale subplot
plt.subplot(1, 2, 2)
plt.boxplot(df['Sales'])
plt.yscale('log')
plt.title('Logarithmic Y Scale')
plt.show()
```

### Histogram

Histograms are useful for visualizing distribution of single varibles

```python
# Histogram of sales data (for info y-axis gives us Count and the x-axis gives us the range)
plt.hist(df['Sales'])
plt.show()
```

```python
# this histogram can be made more readable by using a Log scale as above

plt.hist(df['Sales'])
plt.yscale('log')
plt.show()
```

### Scatter Plot

Scatter plots are used to visualize two variables one on each axis

```python
# Scatter plot with two variables: Profit and Sales
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```
