# Machine Learning

## Types of Machine Learning

Machine Learning models can be classified into **Supervised Learning** and **Unsupervised Learning**, and these types are based on tasks performed and the nature of the output.

## What is Supervised and Unsupervised Learning

Lets's take an example: Suppose we have Lego blocks, we are teaching a kid how to arrange the blocks to create some structures out of them. We built 100 structures of houses of different shapes, cars and ships each. We show them to the kid and tell him what is a house, a car and what is a ship and their properties. After a while he will start to identify the different structutes, their properties on his own. Now, if we build some new structure he will correctly identify what the structure is based on the past experience. This is known as ****Supervised Learning****. Here the 'Label' is known.

Let's twist the above example a little bit, this time we built the structures but we don't explicitly explain the kid that what every sturcture is! He is told to identify the structures on his own. He will do so, based on the properties of the structure, but with less accuracy. It will take a few iterations for him to be able to successfully differentiate between them. But he still doesn't know the names of the different structures and what to call them. This is known as ****Unsupervised Learning****. Here the 'Label' is unknown.

### Supervised Learning

#### 1. Regression
In **Regression** the expected output value is going to be **Continuous variable**.
***Example***: If we want to predict marks of a student in an examination based on that students previous marks he achieved in his prior exams. *marks are continuous in nature (numerical)*.
also, another example, if we want to predict weather or temperature using prior data available, or predict height of some people using weight data of those people.

#### 2. Classification
In **Classification** the output variable which we are going to predict is **Categorical variable**.
Categorical variables are like discrete variables.
***Example***: E-mail classifiers are one example, most e-mail services have classifier systems, which classify which emails in your inbox are legit email, or they are spam mails, there are only two possibilities (spam or not spam).
These kind of problems are called Classificational algorithms.

### Unsupervised Learning

#### 3. Clustering
In both **Regression** and **Classification** we have the output variable/column with us (also called as **"Label"**.
In **Clustering** we do not have any sort of label with us, here we will try to look at some properties  without any knowledge of label and may be we would try to group them together, things having similar properties might get grouped together. This lond of problem will fall into the Clustering category.
***Example***: Google News aggregator service is one example of clustering, this service wiil aggregate news feeds which have similar topics or genre or related some how.
